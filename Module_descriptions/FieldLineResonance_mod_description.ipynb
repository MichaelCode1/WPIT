```python
#import required packages
import numpy as np
import os
import sys
import matplotlib.pyplot as plt


#Define WPIT package location
current_dir =  os.path.abspath(os.path.dirname('__file__'))
fpath = os.path.abspath(current_dir + "/..")
#Add WPIT path to Python path
sys.path.append(fpath)

#import WPIT modules
import WPIT.Environment_mod as env
import WPIT.WaveProperties_mod as wave
import WPIT.WPI_mod.EMIC_ion_mod as wpi
import WPIT.FLR_mod as flr
```

```python
#define constants
m=0
phi=0
Bm=50.0e-9
BE=3.11e-5
L=5.7
sm=(1-1/L)**(1/2)
f0=1e-2
om0=2*np.pi*f0
Va0=om0*L*env.const.Re/flr.fN(L)
omN=Va0*flr.fN(L)/(L*env.const.Re)
Beq=BE/L**3
```

```python
# radial field in nT
t = 0    # Example value for t, replace with actual value

# Generate s values
s_values = np.linspace(-sm, sm, 100)

# Calculate len(s)/RE and B2(s, t) * 10^9
x_values = flr.FL_len(L, s_values) / env.const.Re
y_values = flr.B2(L, s_values, t, f0, Bm) * 1e9  # Convert to 10^9

# Plot the parametric curve
plt.plot(x_values, y_values)

# Set plot range
# plt.xlim(min(x_values), max(x_values))  # Adjust x-axis limits if needed
# plt.ylim(min(y_values), max(y_values))  # Adjust y-axis limits if needed

# Show the plot
#plt.figure(figsize=(8, 6))
plt.xlabel('Length / RE')
plt.ylabel('B2 (T)')
plt.title('B field for Field Line in nT') 

plt.grid(True)

plt.show()
```

```python
# azimuthal field in mV/m

t = np.pi/2 * 1/(om0+((omN**2-om0**2)/(4*om0)))

#x_values = length(s_values) / RE
y_values = flr.E3(L, s_values, t, f0, Bm) * 1e3 # Convert to 10^9

# Plot the parametric curve
plt.plot(x_values, y_values)

# Set plot range
# plt.xlim(min(x_values), max(x_values))  # Adjust x-axis limits if needed
# plt.ylim(min(y_values), max(y_values))  # Adjust y-axis limits if needed

# Show the plot
plt.xlabel('Length / RE')
plt.ylabel('E3 (mV/m)')
plt.title('Azimuthal Field in mV/m')

plt.grid(True)

plt.show()
```

```python
# compressional field in nT
t = 0
s_values = np.linspace(-sm, sm, 1000)
#x_values = length(s_values) / RE
x_values = flr.FL_len(L, s_values) / env.const.Re
y_values = flr.B11(L, s_values, t, f0, Bm) * 1e9 # Convert to 10^9

# Plot the parametric curve
plt.plot(x_values, y_values)

# Show the plot
plt.xlabel('length[s] / RE')
plt.ylabel('B11(s, t=0) (nT)')
plt.title('compressional field in nT')

plt.grid(True)

plt.show()
```
